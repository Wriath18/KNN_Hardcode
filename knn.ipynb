```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv("xdata.csv")
Y = pd.read_csv("ydata.csv")
```

```python
X = X.drop(["0"], axis=1)
Y = Y.drop(["0"], axis=1)
```

```python
X_data = X.values
Y_data = Y.values
plt.scatter(X_data[:, 0], X_data[:, 1], c=Y_data)
X_data[0]
```

```python
def distance_formula(point1, point2):
    return (((point2[0] - point1[0])**2) + ((point2[1] - point1[1])**2))**0.5
```

```python
class KNN:
    def __init__(self, k=5):
        self.k = k

    def fitting(self, X, Y):
        self.X = X
        self.Y = Y

    def distance_formula(self, p1, p2):
        return (((p1[0] - p2[0])**2) + ((p1[1] - p2[1])**2))**0.5
    
    def predict_class(self, p):
        distances = []
        for i in range(len(self.X)):
            distances.append((self.distance_formula(self.X[i], p), i))

        distances = sorted(distances, key=lambda x:x[0])
        points_with_distances = distances[:self.k]

        classes = []
        for distance, index in points_with_distances:
            classes.append(self.Y[index])

        classs, counts = np.unique(classes, return_counts=True)
        
        return classes[np.argmax(counts)]
```

```python
model = KNN()
model.fitting(X_data, Y_data)
val = model.predict_class(X_data[263])
print(val)
```
